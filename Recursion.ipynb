For Loop vs Recursion

```python
for i in range (1,11):
        print(i , end="   ")
```

1) Head Recursion

```python
def rec(s,n) : 
    if s==n:
        return

    print(s , end="    ")
    return rec(s+1,n)
```

```python
rec(1,11)
```

2) Tail Recursion

```python
def rec(s,n) : 
    if s==n:
        return
    
    rec(s+1,n)
    print(s , end="   ")
```

```python
rec(1,11)
```

Practice Question Factorial

```python
n = 5
fact = 1

for i in range(1, n + 1):
    fact = fact * i

print("Factorial =", fact)
```

```python
def fact(n):
    if n==1 :
        return n
    return n*fact(n-1)
```

```python
fact(5)
```
